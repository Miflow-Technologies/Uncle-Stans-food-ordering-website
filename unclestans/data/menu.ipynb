```python
import pandas as pd
```

```python
df = pd.read_csv('menu.csv')
df.head(10)
```

```python
df_grouped_category = df.groupby('category')
df_grouped_category
```

```python
for group, group_df in df_grouped_category:
    print('Group:', group)
    print(group_df)
```

```python


# Group the data by 'category'
grouped = df.groupby('category')

# Create a dictionary to store the groups
group_dict = {}

# Iterate over the groups
for group, group_df in grouped:
    # Add the group DataFrame to the dictionary
    group_dict[group] = group_df

# Access and print a specific group DataFrame
print(group_dict)
```

```python
packs = group_dict['packs']
platter5 = group_dict['platter for 5']
platter10 = group_dict['platter for 10']
extras = group_dict['extras']
```
